```python
#Import all the necessary libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data
df=pd.read_csv("tendulkar_ODI.csv")
df
```

```python
#Check dataframe shape
df.shape
```

```python
#checking column names
df.columns
```

```python
#Renaming column 4s to 'Fours'
df = df.rename({'4s': 'Fours'}, axis=1)
```

```python
# Clean Quality Issues in Runs Column
#df['Runs'] = df['Runs'].str.replace(r'\D', '', regex=True)
#Removing * and '-' rows
df['Runs'] = [x.rstrip('*') for x in df['Runs']]
df = df[df.Fours != '-']
```

```python
df.shape
```

```python
# Getting Runs Frequency
Freq = df['Runs'].value_counts()
```

```python
df.to_csv('tendulkar_ODI_cleaned.csv', index=False)
```

```python
Freq.plot(kind='bar', title='Frequency Count of Runs', xlabel='Runs', ylabel='Freq')
plt.gcf().set_size_inches(20, 6)
plt.show()
```

```python
df['Fours'] = pd.to_numeric(df['Fours'])
```

```python
# Plot a histogram of a single column in the DataFrame
df.hist(column='Fours')

# Set the title and axis labels
plt.title('Histogram of Fours')
plt.xlabel('Fours')
plt.ylabel('Freq')

# Display the histogram
plt.show()
```
